```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import pandas as pd

# Data for network structure
elements = [
    ("C1", 1, 2), ("C2", 2, 3), ("C3", 3, 4), ("C4", 4, 5),
    ("C5", 5, 6), ("C6", 6, 7), ("C7", 4, 8), ("C8", 8, 9),
    ("C9", 9, 10), ("D1", 4, "ST")
]

node_data = {
    1: {"label": "HW", "elevation": 540.0},
    10: {"label": "FB1", "elevation": 500.0},
    7: {"label": "FB2", "elevation": 500.0},
    "ST": {"label": "Storage Tank", "elevation": 100.0}
}

# Computational Parameters
DTCOMP = 0.01
DTOUT = 0.1
TMAX = 500.0

# Pipe data (length, diameter, Manning's coefficient)
pipe_data = [
    (1000, 0.5, 0.01), (500, 0.4, 0.01), (400, 0.3, 0.012),
    (300, 0.3, 0.012), (300, 0.25, 0.013), (200, 0.2, 0.014),
    (200, 0.2, 0.014), (100, 0.15, 0.015), (100, 0.15, 0.015)
]

def calculate_head_loss(length, diameter, manning_coefficient, flow_rate):
    area = np.pi * (diameter / 2) ** 2  # Cross-sectional area (m^2)
    hydraulic_radius = diameter / 4  # Hydraulic radius (m)
    velocity = flow_rate / area  # Velocity (m/s)
    head_loss = (
        (manning_coefficient ** 2) * (length / hydraulic_radius ** (4 / 3)) * (velocity ** 2) / (2 * 9.81)
    )
    return head_loss

# Initialize time steps and outputs
time_steps = np.arange(0, TMAX + DTCOMP, DTCOMP)
time_out = np.arange(0, TMAX + DTOUT, DTOUT)
flow_rates = []
node_10_head = []
node_3_pressure = []
node_3_head = []

# Simulation loop
for t in time_steps:
    if t < TMAX / 2:  # Ramp-up phase
        flow_rate = 100 * (t / (TMAX / 2))
    else:  # Ramp-down phase
        flow_rate = 100 * ((TMAX - t) / (TMAX / 2))

    flow_rates.append(flow_rate)

    # Compute head at Node 10
    current_pressure = node_data[1]["elevation"] * 1000 * 9.81
    pressures = [current_pressure]

    for length, diameter, manning_coeff in pipe_data:
        head_loss = calculate_head_loss(length, diameter, manning_coeff, flow_rate)
        current_pressure -= head_loss * 1000 * 9.81
        pressures.append(current_pressure)

    node_10_head.append(pressures[-1] / (1000 * 9.81))
    node_3_pressure.append(pressures[2])
    node_3_head.append(pressures[2] / (1000 * 9.81))

# Extract results at output intervals
output_results = {
    "Time (s)": time_out,
    "Node 10 Flow (Q)": [flow_rates[int(t / DTCOMP)] for t in time_out],
    "Node 10 Head": [node_10_head[int(t / DTCOMP)] for t in time_out],
    "Node 3 Pressure": [node_3_pressure[int(t / DTCOMP)] for t in time_out],
    "Node 3 Head": [node_3_head[int(t / DTCOMP)] for t in time_out]
}
output_df = pd.DataFrame(output_results)

# Plotting network diagram
G = nx.DiGraph()
for element_id, start, end in elements:
    G.add_edge(start, end, label=element_id)

# Define positions for plotting (manually adjusting for clarity)
positions = {
    1: (0, 2), 2: (1, 2), 3: (2, 2), 4: (3, 2), 5: (4, 3),
    6: (5, 2), 7: (5, 1), 8: (4, 1), 9: (5, 0), 10: (6, 0),
    "ST": (3, 3)
}

plt.figure(figsize=(12, 8))
nx.draw(
    G, positions, with_labels=True, node_size=700, node_color="lightblue",
    labels={node: f"{node} ({node_data.get(node, {}).get('label', '')})" for node in G.nodes}
)
nx.draw_networkx_edge_labels(
    G, positions, edge_labels=nx.get_edge_attributes(G, "label"), font_color="red"
)
plt.title("Network Flow Diagram")
plt.show()

# Plotting results
plt.figure(figsize=(10, 6))
plt.plot(output_df["Time (s)"], output_df["Node 10 Flow (Q)"], label="Node 10 Flow (Q)", color="green")
plt.xlabel("Time (s)")
plt.ylabel("Flow Rate (m^3/s)")
plt.title("Flow Rate at Node 10")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(output_df["Time (s)"], output_df["Node 3 Pressure"] / (1000 * 9.81), label="Node 3 Pressure (m)", color="blue")
plt.plot(output_df["Time (s)"], output_df["Node 3 Head"], label="Node 3 Head (m)", color="red")
plt.xlabel("Time (s)")
plt.ylabel("Pressure / Head (m)")
plt.title("Pressure and Head at Node 3")
plt.legend()
plt.grid(True)
plt.show()

# Output results as a table
print(output_df)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import pandas as pd
from google.colab import files

# Function to load data from the Excel file
def load_data_from_excel(file_path="/content/Problem.xlsx"):
    # Read all the relevant sheets from the Excel file
    node_data = pd.read_excel(file_path, sheet_name='Node No.')
    conduit_data = pd.read_excel(file_path, sheet_name='Conduit Data')
    surge_tank_data = pd.read_excel(file_path, sheet_name='Surge Tank Properties')
    orifice_data = pd.read_excel(file_path, sheet_name='Orifice Properties')
    flow_schedule = pd.read_excel(file_path, sheet_name='Flow Schedule')
    computational_params = pd.read_excel(file_path, sheet_name='Computational Parameters')

    return node_data, conduit_data, surge_tank_data, orifice_data, flow_schedule, computational_params

# Upload file in Google Colab
uploaded = files.upload()

# After upload, get the file name
file_path = list(uploaded.keys())[0]  # This will pick the first uploaded file

# Load data from the uploaded Excel file
node_data, conduit_data, surge_tank_data, orifice_data, flow_schedule, computational_params = load_data_from_excel(file_path)

# Convert node elevation from meters to feet (if needed)
node_data['Elevation'] = node_data['Elevation'] * 3.28084  # Convert from meters to feet

# Computational Parameters (from the Excel file)
DTCOMP = computational_params.loc[computational_params['Parameter'] == 'DTCOMP', 'Value'].values[0]
DTOUT = computational_params.loc[computational_params['Parameter'] == 'DTOUT', 'Value'].values[0]
TMAX = computational_params.loc[computational_params['Parameter'] == 'TMAX', 'Value'].values[0]

# Extracting node data into a dictionary
node_data_dict = {
    row['Node No.']: {'label': row['Label'], 'elevation': row['Elevation']}
    for _, row in node_data.iterrows()
}

# Prepare the network structure (elements from conduit data)
elements = [
    (row['Element ID'], row['Length (m)'], row['Dia (m)'], row['Friction - Mannings Coefficient, n'])
    for _, row in conduit_data.iterrows()
]

# Function to calculate head loss (same as before)
def calculate_head_loss(length, diameter, manning_coefficient, flow_rate):
    area = np.pi * (diameter / 2) ** 2  # Cross-sectional area (ft^2)
    hydraulic_radius = diameter / 4  # Hydraulic radius (ft)
    velocity = flow_rate / area  # Velocity (ft/s)
    head_loss = (
        (manning_coefficient ** 2) * (length / hydraulic_radius ** (4 / 3)) * (velocity ** 2) / (2 * 32.174)
    )
    return head_loss

# Extract flow schedule data
time_steps = flow_schedule['Time (s)'].values
flow_rates = flow_schedule['Discharge (cumecs)'].values

# Initialize time steps and outputs
time_out = np.arange(0, TMAX + DTOUT, DTOUT)
node_10_head = []
node_3_pressure = []
node_3_head = []

# Simulation loop
for t in time_steps:
    # Extract the flow rate for this time step
    flow_rate = flow_rates[int(t / DTCOMP)] if t < len(flow_rates) else 0

    # Compute head at Node 10
    current_pressure = node_data_dict[1]['elevation'] * 62.4  # Convert to lb/ft^2
    pressures = [current_pressure]

    for length, diameter, manning_coeff in elements:
        head_loss = calculate_head_loss(length, diameter, manning_coeff, flow_rate)
        current_pressure -= head_loss * 62.4
        pressures.append(current_pressure)

    node_10_head.append(pressures[-1] / 62.4)
    node_3_pressure.append(pressures[2])
    node_3_head.append(pressures[2] / 62.4)

# Output results as a dataframe
output_results = {
    "Time (s)": time_out,
    "Node 10 Flow (Q)": [flow_rates[int(t / DTCOMP)] for t in time_out],
    "Node 10 Head (ft)": [node_10_head[int(t / DTCOMP)] for t in time_out],
    "Node 3 Pressure (psf)": [node_3_pressure[int(t / DTCOMP)] for t in time_out],
    "Node 3 Head (ft)": [node_3_head[int(t / DTCOMP)] for t in time_out]
}
output_df = pd.DataFrame(output_results)

# Plotting network diagram
G = nx.DiGraph()
for _, start, end in elements:
    G.add_edge(start, end)

# Define positions for plotting (manually adjusting for clarity)
positions = {node: (i, 0) for i, node in enumerate(node_data_dict.keys())}

plt.figure(figsize=(12, 8))
nx.draw(
    G, positions, with_labels=True, node_size=700, node_color="lightblue",
    labels={node: f"{node} ({node_data_dict.get(node, {}).get('label', '')})" for node in G.nodes}
)
plt.title("Network Flow Diagram")
plt.show()

# Plotting results
plt.figure(figsize=(10, 6))
plt.plot(output_df["Time (s)"], output_df["Node 10 Flow (Q)"], label="Node 10 Flow (Q)", color="green")
plt.xlabel("Time (s)")
plt.ylabel("Flow Rate (ft^3/s)")
plt.title("Flow Rate at Node 10")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(output_df["Time (s)"], output_df["Node 3 Pressure (psf)"] / 144, label="Node 3 Pressure (psi)", color="blue")
plt.plot(output_df["Time (s)"], output_df["Node 3 Head (ft)"], label="Node 3 Head (ft)", color="red")
plt.xlabel("Time (s)")
plt.ylabel("Pressure / Head")
plt.title("Pressure and Head at Node 3")
plt.legend()
plt.grid(True)
plt.show()

# Output results as a table
print(output_df)
```
